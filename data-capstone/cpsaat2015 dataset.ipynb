```python
import pandas as pd
import numpy as np
import math
import re
```

```python
df = pd.read_csv("cpsaat2015.csv")
```

```python
inner_index = df['Unnamed: 1'].to_numpy()
outer_index = np.array(df['Unnamed: 0'])
columns = pd.Series(list(df.keys())[2:-2])
data = df[columns].values
```

```python
for i in range(0, len(outer_index)):
    try:
        if re.findall('[a-z]', outer_index[i]):
            new_word = outer_index[i]
        elif new_word != outer_index[i+1]:
            new_word = outer_index[i+1]
    except:
        outer_index[i] = new_word
```

```python
index_array = [outer_index, inner_index]
```

```python
pd.DataFrame(data, index=index_array, columns=columns)
```
