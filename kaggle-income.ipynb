```python
import numpy as np
import pandas as pd
import math
import re
```

```python
to_clean = []
with open(r'kaggle_income.csv') as f:
    for lines in f:
        to_clean.append(lines.replace('\n', '').replace('_', ' ').title())
```

```python
columns = to_clean[0].split(',')
```

```python
clean_data = []
for i in range(1, len(to_clean)):
    raw_data = to_clean[i]
    clean_data.append(raw_data.split(','))
```

```python
clean_data
```

```python
df = pd.DataFrame(clean_data, columns=columns)
```

```python
columns = ['Aland', 'Awater', 'Mean', 'Median', 'Stdev', 'Sum W']
df[columns] = df[columns].astype(float)
```
